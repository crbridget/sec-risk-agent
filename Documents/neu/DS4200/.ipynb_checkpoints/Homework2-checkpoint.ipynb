The Xenographics project is a curated collection of unusual, experimental, and innovative data visualizations created by Maarten Lambrechts in 2017, with the goal of inspiring data visualization practitioners, fighting "xenographphobia" (the fear of unfamiliar charts), and popularizing novel chart types that might be effective for specific datasets. As Lambrechts explains in his OpenVis talk, we are living in an era similar to William Playfair's time, when fundamentally different data requires fundamentally different visualizations, and he argues that we need "xenographics" (strange or weird graphics that people haven't seen before) to effectively communicate today's complex, multidimensional datasets. The visualizations on xeno.graphics are organized using a tagging system that categorizes charts by data aspects and analytical purposes, including correlation, distribution, domain-specific, geo, hierarchy, high-dimensional, mashup, network, scale stacking, time, flow, sets, uncertainty, and mirroring, allowing users to filter and discover appropriate visualization techniques for their specific data challenges.

The most surprising insight from Lambrechts' talk was his discussion of the "xenographic catch-22"where publishers refuse to publish complex visualizations because audiences have low graphic literacy, yet audiences maintain low graphic literacy precisely because they're never exposed to more sophisticated charts. His argument that we must break this cycle by publishing weirder charts with clear explanations resonated strongly, especially his invocation of Playfair's confidence that "with reading a few lines of directions all the difficulty...entirely vanishes." My favorite visualization is the OD Map <https://xeno.graphics/articles/od-map/>, created by Jo Wood, which visualizes origin-destination flow data through a fractal-like arrangement of small multiple maps. It's simultaneously disorienting and brilliant, showing how geographic flows can be represented without the visual clutter of traditional flow maps.

### Part 3 - Basic data analysis with Altair (40 points)

We are going to use the same Weather data from Homework 1 but use Altair to create the data visualziation. 

##### Question 1 (10 points)

Recreate the figure in Part 3 Question 2 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and weather type.

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import altair as alt
from vega_datasets import data

weather = data.seattle_weather()
weather.head()
```

```python
chart = alt.Chart(weather).mark_boxplot().encode(
    x=alt.X('weather:N', title='weather type'),
    y=alt.Y('temp_max:Q', title='maximum temperature'),
    color='weather:N'
).properties(
    title='Seattle Weather - Maximum Temperature by Weather Type',
    width=400,
    height=300
)

chart
```

##### Question 2 (10 points)

Recreate the figure in Part 3 Question 3 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type.

```python
chart = alt.Chart(weather).mark_circle(opacity=0.6, size=60).encode(
    x=alt.X('temp_max:Q', title='max temperature'),
    y=alt.Y('temp_min:Q', title='min temperature'),
    color=alt.Color('weather:N',
                    scale=alt.Scale(range=['#4C72B0', '#DD8452', '#55A868', '#C44E52', '#8172B3']))
).properties(
    title='The Relationship Between Max and Min Temp (Seattle)',
    width=500,
    height=400
)

chart
```

##### Question 3 (10 points)

Create a facet figure in Altair with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python
chart = alt.Chart(weather).mark_circle(size=60).encode(
    x=alt.X('temp_max:Q', 
            title='Maximum Temperature (F)',
            scale=alt.Scale(zero=False)),
    y=alt.Y('temp_min:Q', 
            title='Minimum Temperature (F)',
            scale=alt.Scale(zero=False)),
    color=alt.Color('wind:Q', 
                    title='Wind Speed',
                    scale=alt.Scale(scheme='viridis')),
    tooltip=[
        alt.Tooltip('temp_max:Q', title='Max Temp', format='.1f'),
        alt.Tooltip('temp_min:Q', title='Min Temp', format='.1f'),
        alt.Tooltip('wind:Q', title='Wind Speed', format='.1f')
    ],
    facet=alt.Facet('weather:N', columns=3, title='Weather Type')
).properties(
    width=200,
    height=200,
    title='Temperature Relationship by Weather Type (Seattle)'
).resolve_scale(
    color='independent'
)

chart
```

##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Note, just use the `date` as X attribute, no need to aggragate for years and you will get a wide figure. Include necessary title and labels.

```python
# Calculate average temperature
weather['temp_avg'] = (weather['temp_max'] + weather['temp_min']) / 2

# Base chart
base = alt.Chart(weather).encode(
    x=alt.X('date:T', title='Date')
)

# Line chart for the trend
line = base.mark_line(color='steelblue', size=2).encode(
    y=alt.Y('temp_avg:Q', title='Average Temperature (F)')
)

# Points to show individual data points
points = base.mark_circle(size=30, opacity=0.3, color='steelblue').encode(
    y='temp_avg:Q',
    tooltip=[
        alt.Tooltip('date:T', title='Date', format='%Y-%m-%d'),
        alt.Tooltip('temp_avg:Q', title='Avg Temp', format='.1f'),
        alt.Tooltip('temp_max:Q', title='Max Temp', format='.1f'),
        alt.Tooltip('temp_min:Q', title='Min Temp', format='.1f')
    ]
)

# Combine the charts
chart = (line + points).properties(
    width=800,
    height=400,
    title='Average Temperature Trend Over Time (Seattle)'
)

chart
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Create figure
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.set_xlim(0, 12)
ax.set_ylim(0, 6)
ax.axis('off')

# Define the identical gray color for both central squares
identical_gray = '#808080'

# Left side: Gray square on dark background
dark_background = patches.Rectangle((0.5, 0.5), 5, 5, 
                                    linewidth=0, 
                                    facecolor='#1a1a1a')
ax.add_patch(dark_background)

gray_square_left = patches.Rectangle((2, 2), 2, 2, 
                                     linewidth=0, 
                                     facecolor=identical_gray)
ax.add_patch(gray_square_left)

# Right side: Same gray square on light background
light_background = patches.Rectangle((6.5, 0.5), 5, 5, 
                                     linewidth=0, 
                                     facecolor='#e8e8e8')
ax.add_patch(light_background)

gray_square_right = patches.Rectangle((8, 2), 2, 2, 
                                      linewidth=0, 
                                      facecolor=identical_gray)
ax.add_patch(gray_square_right)

# Add labels
ax.text(3, 0.2, 'Dark Background', ha='center', fontsize=12, fontweight='bold')
ax.text(9, 0.2, 'Light Background', ha='center', fontsize=12, fontweight='bold')

# Add title
plt.title('Simultaneous Contrast Illusion: Both Gray Squares Are Identical (#808080)', 
          fontsize=14, fontweight='bold', pad=20)

# Add annotation
ax.text(6, 5.5, 'The left square appears lighter\nThe right square appears darker', 
        ha='center', va='center', fontsize=11, 
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.savefig('simultaneous_contrast_illusion.png', dpi=300, bbox_inches='tight')
plt.show()
```
